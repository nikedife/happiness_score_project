# Dropping columns without much correlation with the target

```python
X = df.drop(['Standard Error','Generosity','Country','Region','Happiness Score','Happiness Rank'],axis=1)
```

```python
y = df['Happiness Score']
```

# Lets build our model and make some predictions

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,adjusted_rand_score
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
lr = LinearRegression()
```

```python
X_scaled = scale.fit_transform(X)
```

```python
for i in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=i)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    if (r2_score(y_train,pred_train) and r2_score(y_test,pred_test))>0.996: 
        print(f"At random state of {i}, the training accuracy is:- {r2_score(y_train,pred_train)}")
        print(f"At random state of {i}, the testing accuracy is:- {r2_score(y_test,pred_test)}")
        print("\n")


    
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_scaled,y,test_size=0.2,random_state=45)

lr = LinearRegression()
lr.fit(x_train,y_train)
pred_test = lr.predict(x_test)
```

# Lets evaluate our model and check for the error too

```python
val_test = r2_score(y_test,pred_test)
print(val_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

print('mean adbsolute error:',mean_absolute_error(y_test,pred_test))
print('mean squared error:',mean_squared_error(y_test,pred_test))
print('Root mean squared error:',np.sqrt(mean_squared_error(y_test,pred_test)))
```

```python
train_accuracy = r2_score(y_train,pred_train)
test_accuracy = r2_score(y_test,pred_test)

from sklearn.model_selection import cross_val_score

for j in range(2,10):
    cv_score = cross_val_score(lr,X_scaled,y,cv=j)
    cv_mean = cv_score.mean()
    print(f"at cv {j} the cv score is {cv_mean} and accuracy score for training is {train_accuracy} and accuracy score for testing is {test_accuracy}")
    print("\n")
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x=y_test,y=pred_test,color='r')
plt.plot(y_test,y_test,color='b')
plt.xlabel('actual charges')
plt.ylabel('predicted charges')
plt.title('linear regression')
plt.savefig('lr.png')
plt.show()
```

# Lets save our model and make some predictions and compare with the actual figures

```python
import pickle
filename = 'happiness.pkl'
pickle.dump(lr,open(filename,'wb'))
```

```python
loaded_model = pickle.load(open('happiness.pkl','rb'))
result = loaded_model.score(x_test,y_test)
print(result*100)
```
